```python
import numpy as np
import matplotlib.pyplot as plt

def fill_with_sine(num_weights, vmin, vmax):
    # Create an array of equally spaced points between vmin and vmax
    x_values = np.linspace(vmin, vmax, num=num_weights)
    
    # Compute the sine of each element in the array
    sine_values = np.sin(x_values)
    
    return sine_values

# Example usage:
num_weights = 31
vmin = 0
vmax = np.pi * 2
distance_per_weight = (vmax - vmin) / (num_weights) 

weights = fill_with_sine(num_weights, vmin, vmax - distance_per_weight)
# weights = fill_with_sine(num_weights, vmin, vmax)
# weights = np.linspace(0, 10, num_weights)
start = -10
end = 10
num = 1000
values = np.linspace(start, end, num=num)
x = np.linspace(start, end, num=num)
# values = np.linspace(0, 10, 10)
# x = values
distance = vmax - vmin
# weights[0] = 1

values[values < vmin] = vmax - ((vmin - values[values < vmin]) % distance)
values[values > vmax] = ((values[values > vmax] - vmax) % distance) + vmin



num_distance = (values - vmin) / distance_per_weight

lower_weight_index = np.floor(num_distance).astype(int)
upper_weight_index = np.ceil(num_distance).astype(int) % num_weights

values = weights[lower_weight_index] + (num_distance - lower_weight_index) * (weights[upper_weight_index] - weights[lower_weight_index])

print(weights)

plt.plot(x, values, marker='*', color='blue')
plt.xlabel('x')
plt.ylabel('Values')
plt.title('Values vs. x')
plt.grid(True)
plt.show()
```
